## Challenge 1 - Who Have Published What At Where?


```python
query = '''

SELECT
    authors.au_id,
    authors.au_lname,
    authors.au_fname,
    titles.title,
    publishers.pub_name
    
FROM
    `ironhack-data-analytics.publications.authors`authors
        
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor`titleauth

ON
    authors.au_id = titleauth.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles`titles
    
ON    
    titleauth.title_id = titles.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.publishers`publishers

ON
    titles.pub_id = publishers.pub_id
       

ORDER BY
publishers.pub_id


'''
query_job = client.query(query=query)
df=query_job.to_dataframe()
df
```

## Challenge 2 - Who Have Published How Many At Where?

```python
query = '''

SELECT
    authors.au_id,
    authors.au_lname,
    authors.au_fname,
    publishers.pub_name,
    COUNT(titles.title_id) AS title_count
    
FROM
    `ironhack-data-analytics.publications.authors`authors
        
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor`titleauth

ON
    authors.au_id = titleauth.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles`titles
    
ON    
    titleauth.title_id = titles.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.publishers`publishers

ON
    titles.pub_id = publishers.pub_id

GROUP BY
    1,2,3,4
ORDER BY
    title_count DESC
       



'''
query_job = client.query(query=query)
df=query_job.to_dataframe()
df
```

## Challenge 3 - Best Selling Authors

```python
query = '''

SELECT
    authors.au_id,
    authors.au_lname,
    authors.au_fname,
    SUM(titles.ytd_sales) AS Total
    
FROM
    `ironhack-data-analytics.publications.authors`authors
        
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor`titleauth

ON
    authors.au_id = titleauth.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles`titles
    
ON    
    titleauth.title_id = titles.title_id
    

GROUP BY
    1,2,3
    
ORDER BY
    total DESC
    
LIMIT
    3



'''
query_job = client.query(query=query)
df=query_job.to_dataframe()
df
```

## Challenge 4 - Best Selling Authors Ranking

```python
query = '''

SELECT
    authors.au_id,
    authors.au_lname,
    authors.au_fname,
    SUM(titles.ytd_sales) AS Total
    
FROM
    `ironhack-data-analytics.publications.authors`authors
        
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor`titleauth

ON
    authors.au_id = titleauth.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles`titles
    
ON    
    titleauth.title_id = titles.title_id
    

GROUP BY
    1,2,3
    
ORDER BY
    total DESC
    
LIMIT
    23



'''
query_job = client.query(query=query)
df=query_job.to_dataframe()
df
```

## Bonus Challenge - Most Profiting Authors

```python
query = '''

SELECT
    authors.au_id,
    authors.au_lname,
    authors.au_fname,
    ((titles.ytd_sales * titles.price) * (titles.royalty/100)) as Profit
    
FROM
    `ironhack-data-analytics.publications.authors`authors
        
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor`titleauth

ON
    authors.au_id = titleauth.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles`titles
    
ON    
    titleauth.title_id = titles.title_id
    

GROUP BY
    1,2,3,4
    
ORDER BY
    4 DESC
    
LIMIT
    3
    



'''
query_job = client.query(query=query)
df=query_job.to_dataframe()
df
```
